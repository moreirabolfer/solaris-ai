```python
# ======================================================
# Solaris AI Control
# Sistema Inteligente de Monitoramento Energético Espacial
# Global Solution - Energias Renováveis e Sustentáveis
# ======================================================

def analisar_temperatura(temp):
    if temp > 80:
        return "CRÍTICO", "Ativar sistema de resfriamento"
    elif temp > 60:
        return "ATENÇÃO", "Reduzir consumo energético"
    else:
        return "ESTÁVEL", "Temperatura ideal"


def analisar_energia(energia):
    if energia < 30:
        return "CRÍTICO", "Ativar economia de energia e painéis solares"
    elif energia < 60:
        return "ATENÇÃO", "Monitorar geração renovável"
    else:
        return "ESTÁVEL", "Energia suficiente"


def analisar_comunicacao(sinal):
    if sinal < 40:
        return "CRÍTICO", "Realinhar antenas de comunicação"
    elif sinal < 70:
        return "ATENÇÃO", "Ajustar transmissão de dados"
    else:
        return "ESTÁVEL", "Comunicação funcionando corretamente"



def inteligencia_artificial(temp, energia, comunicacao):

    problemas = 0

    if temp > 60:
        problemas += 1

    if energia < 60:
        problemas += 1

    if comunicacao < 70:
        problemas += 1


    if problemas >= 2:
        return "RISCO ALTO - IA recomenda intervenção imediata"

    elif problemas == 1:
        return "RISCO MODERADO - Continuar monitoramento"

    else:
        return "MISSÃO SEGURA - Sistemas funcionando"



# Dados simulados da missão espacial

dados_missao = [

    {
        "ciclo": 1,
        "temperatura": 45,
        "energia": 95,
        "comunicacao": 98,
        "modulo": "Painéis solares"
    },


    {
        "ciclo": 2,
        "temperatura": 55,
        "energia": 75,
        "comunicacao": 90,
        "modulo": "Controle central"
    },


    {
        "ciclo": 3,
        "temperatura": 68,
        "energia": 55,
        "comunicacao": 75,
        "modulo": "Baterias sustentáveis"
    },


    {
        "ciclo": 4,
        "temperatura": 91,
        "energia": 20,
        "comunicacao": 25,
        "modulo": "Sistema energético principal"
    },


    {
        "ciclo": 5,
        "temperatura": 70,
        "energia": 50,
        "comunicacao": 65,
        "modulo": "Recuperação automática"
    },


    {
        "ciclo": 6,
        "temperatura": 48,
        "energia": 88,
        "comunicacao": 96,
        "modulo": "Sistema estabilizado"
    }

]



print("===== SOLARIS AI CONTROL =====")
print("Monitoramento Inteligente Espacial\n")



for dado in dados_missao:

    print("-----------------------------------")

    print("Ciclo:", dado["ciclo"])

    print("Módulo:", dado["modulo"])


    print("\nDADOS MONITORADOS")

    print("Temperatura:",
          dado["temperatura"],
          "°C")

    print("Energia renovável:",
          dado["energia"],
          "%")

    print("Comunicação:",
          dado["comunicacao"],
          "%")



    temp_status, temp_acao = analisar_temperatura(
        dado["temperatura"]
    )


    energia_status, energia_acao = analisar_energia(
        dado["energia"]
    )


    comunicacao_status, comunicacao_acao = analisar_comunicacao(
        dado["comunicacao"]
    )



    print("\nALERTAS DO SISTEMA")


    print(
        "Temperatura:",
        temp_status,
        "-",
        temp_acao
    )


    print(
        "Energia:",
        energia_status,
        "-",
        energia_acao
    )


    print(
        "Comunicação:",
        comunicacao_status,
        "-",
        comunicacao_acao
    )



    print("\nDECISÃO DA IA:")

    print(
        inteligencia_artificial(
            dado["temperatura"],
            dado["energia"],
            dado["comunicacao"]
        )
    )



print("\nMonitoramento finalizado com sucesso.")
```
